大语言模型： 超大规模参数  +  **预训练**语言模型

架构： Transformer 解码器 <font color=blue>〔 不包含 编码器 ！！！ 〕</font>

* 语言的生成模式：自回归逐词生成文本

----

<font size=4>当我问 DeepSeek 和 Kimi 这个问题：<font color=blue>什么情况下，大语言模型的架构会包含 Transformer 编码器？</font>

大语言模型包含 Transformer 编码器的情况主要取决于任务需求（如文本理解和分析、特征提取）、模型设计目标（如双向理解、长文本处理）以及预训练和微调的需求

如果任务需要**深度理解输入或提取特征**，**编码器**是必不可少的；而对于纯生成任务，通常只需解码器。
