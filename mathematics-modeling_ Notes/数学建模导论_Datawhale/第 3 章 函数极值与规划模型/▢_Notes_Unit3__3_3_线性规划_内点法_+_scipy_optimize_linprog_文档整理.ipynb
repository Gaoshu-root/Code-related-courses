## 3.3 线性规划的 算法原理

### 3.3.1 单纯形法（Simplex Method）

单纯形法：一种**线性规划的求解方法**，通过**不断地在可行域内移动，找到最优解**。


单纯形法的步骤包括如下几个步骤：

>1、确定初始**可行基**和初始基**可行解**，并**建立初始单纯形表**。 <font color=blue>〔 基于规范型规划 〕</font>
>
>2、在当前表的**目标函数对应的行**中，若**所有非基变量的系数 非正**，则得到最优解，算法终止；否则进入下一步。
>
>3、若单纯形表中 $1 \cdots m$ 列构成单位矩阵，在 $j=m+1\cdots n$ 列中，若有某个对应 $x_k$ 的系数列向量 $P_k \leqslant 0$，则停止计算。否则，转入下一步。
>
>4、挑选目标函数对应行中**系数最大的非基变量**作为**进基变量**。
> * 假设 $x_{k}$ 为进基变量，按如下规则计算，可确定 $x_{u}$ 为**出基变量**，转下一步: $~~\theta = \min \Big\{  \frac{b_{i}}{a_{i,k}} ~\Big|~ a_{i,k} > 0 \Big\} = \frac{b_{u}}{a_{u,k}}$
   * 其中 $b_{i}$ 是规范型规划的**常数项**，$a_{i,k}$ 即为在第 $i$ 个约束中变量 $x_k$ 的系数。
>
> 7、以 $a_{u,k}$ 为**主元素**进行迭代，对 $x_k$ 所对应的**列向量**进行如下变换：
* $P_{k} = [a_{1,k}, a_{2,k}, \dots, \color{blue}{a_{u,k}}, \dots, a_{m,k}]^{\top} = [0, 0, \dots, \color{blue}{1}, \dots, 0]^{\top}$
    * 即令向量 $P_{k}$ 除了第 $u$ 个元素为 $1$ 外其他元素都为零。
>
>8、重复 2-5 步，直到**所有检验数非正**后终止，得到最优解。

### 示例 1

问题没有以 linprog 接受的形式呈现。这很容易通过将“大于”不等式约束转换为 **“小于”不等式约束**来补救，方法是两边乘以一个因子 -1。

还要注意，最后一个约束实际上是  $−3≤𝑥_1≤∞$。

最后，对于 $𝑥_0$，我们必须显式地指定边界 $−∞≤𝑥_0≤∞$，因为**默认情况下变量是非负的**。

```python
from scipy.optimize import linprog

# 目标函数： 最小化
# 约束为： ≤ 形式的不等式 或 等式

c = [-1, 4] # 目标函数的参数系数
A = [[-3, 1],
    [1, 2]]  # x0 + 2x1 <= 4  # 约束中 等式或不等式 的参数系数
b = [6, 4]
x0_bounds = (None, None)
x1_bounds = (-3, None)
res = linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds])
print("目标的最小值：", res.fun)
print("最优解 x =", res.x)
print("是否优化成功：", res.message)
```

```python
# The marginals (AKA dual values / shadow prices / Lagrange multipliers) and residuals (slacks) are also available.
# 双重价格
res.ineqlin
```

例如，由于与第二个不等式约束相关的 <font color=blue>marginal</font> 为 -1，如果我们在第二个不等式约束的右侧添加少量 eps，我们预计目标函数的最优值会减少 eps：


```python
eps = 0.05
b[1] += eps  # 右侧添加 0.05
linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds]).fun  # 目标函数的最优值减小 -22 - 0.05
```

同样，由于第一个不等式约束的 <font color=blue>residual</font>  是 39，我们可以在**不影响最优解**的情况下将第一个约束的右侧减小 39。

约束：$-3x_0+x_1 \leq 6$

* $-3x_0+x_1\color{blue}{+39}=6$

```python
b = [6, 4]  # reset to original values
b[0] -= 39  # 第一个约束的右侧减小 39 （residual ）， 不影响最优解
linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds]).fun
```

###  示例 3.3 〔 scipy.optimize.linprog〕

**例 3.3** 求解下面这个简单的规划案例：

$  \begin{cases} \text{minimize}&f = -5x_{1} - x_{2}\\[0.5em] \text{s.t.} & x_{1} + x_{2} \leqslant 5\\ & 2x_{1} + 0.5x_{2} \leqslant 8\\ & x_{1},~~x_{2} \geqslant {0} \end{cases}$

```python
from scipy.optimize import linprog

# 目标函数： 最小化
# 约束为： ≤ 形式的不等式 或 等式

c = [-5, -1]  # # 目标函数的参数系数
A = [[1, 1],
    [2, 0.5]]  # 约束中 等式或不等式 的参数系数
b = [5, 8]
xi_bounds = (0, None)
res = linprog(c, A_ub=A, b_ub=b, bounds=[xi_bounds]*2)
print("目标的最小值：", res.fun)
print("最优解 x =", res.x)
print("是否优化成功：", res.message)
```

```python
res.ineqlin
```

```python
eps = 0.01
b[1] += eps
linprog(c, A_ub=A, b_ub=b, bounds=[xi_bounds, xi_bounds]).fun
```

```python
b = [5, 8]  # reset to original values
b[0] -= 1
linprog(c, A_ub=A, b_ub=b, bounds=[xi_bounds, xi_bounds]).fun
```
