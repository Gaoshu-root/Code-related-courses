### 基本运算

```python
import numpy as np

a = np.array([10, 20, 30, 40])
b = np.arange(4)  # [0, 1, 2, 3]
print("和 a + b 为：", a+b)
print("乘积 a * b 为：", a*b)
print("幂运算 b**2 为：", b**2 )
print("10 * np.sin(a) =", 10*np.sin(a))
print("b < 3 =", b<3)
```

```python
import numpy as np

a = np.array([[1, 1], [0, 1]])
b = np.arange(4).reshape(2, 2)
print("a = ", a)
print("b = ", b)

print("矩阵乘法 ab =", np.dot(a,b))
print("矩阵乘法 ab =", a.dot(b))
```

```python
import numpy as np

a = np.random.random((2, 4))
print("a = ", a)
print("矩阵元素求和为：", np.sum(a))
print("矩阵元素的最小值为：", np.min(a))
print("矩阵元素的最大值为：", np.max(a))
print("每行 元素分别求和得：", np.sum(a, axis=1))
print("每列 元素分别求和得：", np.sum(a, axis=0))
print("每列 分别获取最小值：", np.min(a, axis=0))
print("每行 分别获取最大值：", np.max(a, axis=1))
```

`numpy.argmin(a, axis=None, out=None, *, keepdims=<no value>)[source]`

沿轴返回最小值的索引。

* By default, the index is into the flattened array, otherwise along the specified axis.
* 默认情况下，索引指向扁平数组，否则沿着指定的轴。

https://numpy.org/doc/stable/reference/generated/numpy.argmin.html#numpy-argmin


numpy.argmax 类似：

https://numpy.org/doc/stable/reference/generated/numpy.argmax.html#numpy.argmax





```python
import numpy as np

A = np.array([[ 5, 3, 4, 200], [ 6, 7, 8, 9], [10, 11, 12, 2]])
print("A =", A)
print(np.argmin(A))  # 一共 12 个元素
print("每列最小值的索引：", np.argmin(A, axis=0))
print("每行最小值的索引：", np.argmin(A, axis=1))
```

```python
print("全部元素平均值：", np.mean(A))
print("全部元素平均值：", np.average(A))
print("元素累计和：", np.cumsum(A))
```

`numpy.diff(a, n=1, axis=-1, prepend=<no value>, append=<no value>)`


* Calculate the n-th discrete difference along the given axis.
* 计算沿给定轴的第 n 个离散差分。

The axis along which the difference is taken, default is the last axis.

取差值的轴，默认是最后一个轴。

https://numpy.org/doc/stable/reference/generated/numpy.diff.html#numpy.diff

```python
print(np.diff(A))  #
```

```python
import numpy as np

A = np.arange(14,2, -1).reshape((3,4))
print("A = ", A)
print("np.sort(A)：", np.sort(A)) # 默认 sort along the last axis
print("A 的转置：", np.transpose(A))
print("A 的转置：", A.T)
```

### 合并

```python
import numpy as np

A = np.array([1, 1, 1])
B = np.array([2, 2, 2])

print("np.vstack((A, B))：\n", np.vstack((A, B))) # 加 行
print("np.hstack((A, B))：\n", np.hstack((A, B))) # 加 列
```

`numpy.concatenate((a1, a2, ...), axis=0, out=None, dtype=None, casting="same_kind")`


Join a sequence of arrays along an existing axis.

沿**现有轴**连接数组序列。


https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html#numpy-concatenate

```python
print("np.concatenate((A, B, B, A), axis=0)：\n", np.concatenate((A, B, B, A), axis=0))
```

```python
import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])

print("np.concatenate((a, b), axis=0):\n", np.concatenate((a, b), axis=0))  #
print("np.concatenate((a, b.T), axis=1):\n", np.concatenate((a, b.T), axis=1))
print("np.concatenate((a, b), axis=None)：\n", np.concatenate((a, b), axis=None))
```

### 分割

https://numpy.org/doc/stable/reference/generated/numpy.split.html#numpy.split

```python
import numpy as np

a = np.arange(12).reshape(3, 4)
print("a:\n", a)

print("")
print("np.split(a, 2, axis=1): \n", np.split(a, 2, axis=1)) # 平均 2

print("")
print("np.split(a, 3, axis=0): \n", np.split(a, 3, axis=0))

print("")
print("np.array_split(a, 3, axis=1): \n", np.array_split(a, 3, axis=1))
```

```python
import numpy as np

x = np.arange(8.0)
print(x)
np.split(x, [3, 5, 6, 10])  # x[:3]  x[3:5] x[5:6] x[6:10] x[10:]
```
