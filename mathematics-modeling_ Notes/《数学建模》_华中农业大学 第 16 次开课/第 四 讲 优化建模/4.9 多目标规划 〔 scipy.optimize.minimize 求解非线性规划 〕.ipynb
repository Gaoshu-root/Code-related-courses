多目标规划的**最优解**： 在给定的偏好关系下， $f$ 在 $X$ 上的最好解。

目标的量纲不一样，且可能互相矛盾。

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html#minimize

from scipy.optimize import minimize

# 目标函数
def func(x):
  x1, x2 = x
  return -0.6 * x1 - 0.7 * x2 + 0.001 * x1**2 + 0.002 * x2**2 + 0.001 *  x1 * x2

# 约束 不等式约束要写成 ≥ 的形式
cons = ({'type': 'eq', 'fun': lambda x:  x[0] + x[1] - 1000},
     {'type': 'ineq', 'fun': lambda x:  x[0] - x[1]})

# 边界
bnds = ((0, None), (0, None))

x_0 = (600, 400) # 猜测的  x 的初始值
res = minimize(func, x_0, method='SLSQP', bounds=bnds,
               constraints=cons)

# print("目标的最小值：", res.fun)  #
print("最大利润为：", [0.6, 0.7] @ res.x)
print("最优解 x =", res.x)
print("是否优化成功：", res.message)
```
